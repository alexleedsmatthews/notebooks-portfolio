```python
df = pd.read_html("https://history.house.gov/People/Office/Speakers-List/")[0]
```

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
```

```python
df["term_start"] = pd.to_datetime(df["date_elected"])
```

```python
df["term_end"] = pd.to_datetime(df["date_elected"].shift(-1))
```

```python
df["days"] = df["term_end"] - df["term_start"]
```

```python
speaker_tenures = (
    df.groupby("name")["days"].sum().reset_index().sort_values("days", ascending=False)
)
```
